```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import requests
from io import BytesIO

r = requests.get('https://docs.google.com/spreadsheets/d/e/2PACX-1vReYTKNET8QeXFNgZyLq6XVnm5PpGOMv1U_HysFpqTevq52fw6wZ1IU4FPYi-SFsMM972oZuw1rzunr/pub?gid=0&single=true&output=csv')
data = r.content
df = pd.read_csv(BytesIO(data))

#df = pd.read_excel('Map_data - Test2.xlsx')

def generateBaseMap(default_location=[56.4971944, 84.9837908], default_zoom_start=12):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map

base_map = generateBaseMap()

free_time = folium.FeatureGroup(name='1. Места для проведения свободного времени / Places for spending free time')
study_time = folium.FeatureGroup(name='2. Места для учебы / Places for studying')
meet_time = folium.FeatureGroup(name='3. Места для встречи с друзьями и студентами / Places to meet friends and students')

df_ru = df.query("Origin == 'Из России / Russian'")
df_ru.reset_index(inplace=True)
df_for = df.query("Origin == 'Иностранный / Foreign'")
df_for.reset_index(inplace=True)

locations1_ru = df_ru[['FreeMap_lat', 'FreeMap_lon']]
locationlist1_ru = locations1_ru.values.tolist()

locations2_ru = df_ru[['StudyMap_lat', 'StudyMap_lon']]
locationlist2_ru = locations2_ru.values.tolist()

locations3_ru = df_ru[['MeetMap_lat', 'MeetMap_lon']]
locationlist3_ru = locations3_ru.values.tolist()

locations1_for = df_for[['FreeMap_lat', 'FreeMap_lon']]
locationlist1_for = locations1_for.values.tolist()

locations2_for = df_for[['StudyMap_lat', 'StudyMap_lon']]
locationlist2_for = locations2_for.values.tolist()

locations3_for = df_for[['MeetMap_lat', 'MeetMap_lon']]
locationlist3_for = locations3_for.values.tolist()


for point in range(0, len(locationlist1_ru)):
    folium.Marker(locationlist1_ru[point], 
                  popup=str(df_ru['FreeMap_desc'][point]) + ', ' + str(df_ru['FreeMap_activity'][point]),
                  icon=folium.Icon(
                      icon='coffee', prefix='fa', color='red')).add_to(free_time)

for point in range(0, len(locationlist1_for)):
    folium.Marker(locationlist1_for[point],
                  popup=str(df_for['FreeMap_desc'][point]) + ', ' + str(df_for['FreeMap_activity'][point]),
                  icon=folium.Icon(
                      icon='coffee', prefix='fa', color='lightblue')).add_to(free_time)
    
for point in range(0, len(locationlist2_ru)):
    folium.Marker(locationlist2_ru[point], 
                  popup=df_ru['StudyMap_desc'][point],
                  icon=folium.Icon(
                      icon='book', prefix='fa', color='red')).add_to(study_time)

for point in range(0, len(locationlist2_for)):
    folium.Marker(locationlist2_for[point], 
                  popup=df_for['StudyMap_desc'][point],
                  icon=folium.Icon(
                      icon='book', prefix='fa', color='lightblue')).add_to(study_time)

for point in range(0, len(locationlist3_ru)):
    folium.Marker(locationlist3_ru[point],
                  icon=folium.Icon(
                      icon='comments-o', prefix='fa', color='red')).add_to(meet_time)

for point in range(0, len(locationlist3_for)):
    folium.Marker(locationlist3_for[point],
                  icon=folium.Icon(
                      icon='comments-o', prefix='fa', color='lightblue')).add_to(meet_time)


free_time.add_to(base_map)
study_time.add_to(base_map)
meet_time.add_to(base_map)

folium.LayerControl().add_to(base_map)


base_map
```
